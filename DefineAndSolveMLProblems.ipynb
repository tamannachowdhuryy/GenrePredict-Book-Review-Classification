```python

# Load the dataset
bookReviewDataSet_filename = os.path.join(os.getcwd(), "data", "bookReviewsData.csv")
df = pd.read_csv(bookReviewDataSet_filename)

# Inspect the data
print("Initial DataFrame:")
print(df.head())

# Check existing columns
print("\nColumns in DataFrame:")
print(df.columns)

# Ensure the 'Review' column exists
if 'Review' not in df.columns:
    raise ValueError("The 'Review' column is missing from the DataFrame.")

# Example genre list (Replace this with actual genre data or method to determine genres)
example_genres = [
    "Fiction", "Non-Fiction", "Mystery", "Science Fiction", "Fantasy",
    "Romance", "Horror", "Historical", "Biography", "Self-Help", "Educational",
    "Thriller", "Adventure", "Dystopian", "Young Adult", "Code",
    "Classics", "Literary Fiction", "Graphic Novel", "Poetry", "Cookbook",
    "Travel", "True Crime", "Science", "Philosophy", "Religion", "Spirituality",
    "Humor", "Politics", "Economics", "Health", "Parenting", "Art", "Music",
    "Reference", "Religion & Spirituality", "Science & Nature", "Sports",
    "Western", "Short Stories", "Contemporary", "Urban Fiction", "Metaphysical",
    "Science Fiction & Fantasy", "Military", "Historical Fiction", "Action",
    "Biographical Fiction", "Inspirational", "Young Adult Fiction", "Cookery",
    "Memoir", "Graphic Memoir", "Self-Improvement", "Mindfulness", "Meditation",
    "Spiritual Fiction", "Supernatural", "Occult", "Mythology", "Adventure Fiction",
    "Literary Non-Fiction", "Philosophical Fiction", "Economic History", "Behavioral Science",
    "Coding", "Programming", "Software Development", "Data Science", "Artificial Intelligence",
    "Machine Learning", "Cybersecurity", "Networking", "Web Development", "Mobile Development",
    "Cloud Computing", "Database Management", "Robotics", "Blockchain", "Game Development",
    "Augmented Reality", "Virtual Reality", "DevOps", "Data Analysis", "Algorithms",
    "Computer Science", "Information Systems", "Statistics", "System Administration",
    "Ethical Hacking", "UX/UI Design", "Hardware", "E-commerce", "Digital Marketing",
    "Product Management", "Tech Startups", "Innovation"
]

# Create a repeating list of genres (if no actual genres provided) or assign genres properly
num_reviews = len(df)
if 'Genre' not in df.columns:
    genres = (example_genres * (num_reviews // len(example_genres) + 1))[:num_reviews]
    df['Genre'] = genres
else:
    print("\n'Genre' column already exists.")
    
print("\nDataFrame with 'Genre' column:")
print(df.head())

# Prepare features (TF-IDF vectors of the reviews)
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Review'])

# Specify the number of clusters
k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Visualize clusters using PCA
pca = PCA(n_components=2)
scatter_plot_points = pca.fit_transform(X.toarray())
colors = ["r", "b", "c", "y", "m"]

x_axis = scatter_plot_points[:, 0]
y_axis = scatter_plot_points[:, 1]
fig, ax = plt.subplots(figsize=(20, 10))

ax.scatter(x_axis, y_axis, c=[colors[d] for d in df['Cluster']])
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Clusters Visualization')
plt.show()

# Assign genres to clusters manually or based on some heuristic (this is an example)
cluster_genres = {
    0: "Science Fiction",
    1: "Romance",
    2: "Mystery",
    3: "Fantasy",
    4: "Non-Fiction"
}

df['Assigned_Genre'] = df['Cluster'].map(cluster_genres)

# Print the DataFrame with the assigned genres
print("\nDataFrame with Assigned Genres:")
print(df.head())

# Prepare features and labels
X = df['Review']
y = df['Assigned_Genre']  # The column with assigned genres

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Create a pipeline with TF-IDF and a classifier
model = make_pipeline(TfidfVectorizer(stop_words='english'), LogisticRegression(max_iter=1000, class_weight='balanced'))

# Train the model
model.fit(X_train, y_train)

# Predict genres for the test set
y_pred = model.predict(X_test)

# Print classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_, zero_division=1))

# Predict genres for the entire dataset
df['Predicted_Genre'] = model.predict(df['Review'])
df['Predicted_Genre'] = label_encoder.inverse_transform(df['Predicted_Genre'])

# Print the DataFrame with the predicted genres
print("\nDataFrame with Predicted Genres:")
print(df.head())  # Adjust to display a subset of the DataFrame as needed
```

```python

# Sample data 
data = {'text': ['sample text data']*100, 'label': ['category']*100}
df = pd.DataFrame(data)

# Prepare data
X = df['text']
y = df['label']

# Encode labels as binary (0 or 1)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

# Create a TfidfVectorizer object
tfidf_vectorizer = TfidfVectorizer()

# Fit the vectorizer to X_train
tfidf_vectorizer.fit(X_train)

# Transform the training and test data
X_train_tfidf = tfidf_vectorizer.transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Print vocabulary size
vocabulary_size = len(tfidf_vectorizer.vocabulary_)
print("Vocabulary size:", vocabulary_size)

# Define the neural network model
nn_model = Sequential()
input_shape = (X_train_tfidf.shape[1],)
nn_model.add(InputLayer(input_shape=input_shape))
nn_model.add(Dense(units=64, activation='relu'))
nn_model.add(Dropout(0.5))  # Added dropout for regularization
nn_model.add(Dense(units=32, activation='relu'))
nn_model.add(Dense(units=16, activation='relu'))
nn_model.add(Dense(units=1, activation='sigmoid'))  # Output layer for binary classification

# Print model summary
nn_model.summary()

# Define the optimizer and loss function
optimizer = Adam(learning_rate=0.001)
loss_fn = 'binary_crossentropy'  # Directly use string for loss function

# Compile the model
nn_model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])

# Define a custom callback to print epoch details
class CustomEpochLogger(Callback):
    def __init__(self, num_epochs: int, every_n: int = 10):
        self.num_epochs = num_epochs
        self.every_n = every_n
    
    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % self.every_n == 0:
            print(f"Epoch [{epoch + 1}/{self.num_epochs}], Loss: {logs['loss']:.4f}, Accuracy: {logs['accuracy']:.4f}, Val_loss: {logs['val_loss']:.4f}, Val_accuracy: {logs['val_accuracy']:.4f}")

# Start time
t0 = time.time()

# Number of epochs
num_epochs = 55

# Fit the model
history = nn_model.fit(X_train_tfidf.toarray(), y_train,
                       epochs=num_epochs,
                       verbose=0,  # Suppress detailed output
                       callbacks=[CustomEpochLogger(num_epochs, every_n=5)],  # Custom callback for logging
                       validation_split=0.2)

# End time
t1 = time.time()
print('Elapsed time: %.2fs' % (t1 - t0))

# Plot training and validation loss
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(range(1, num_epochs + 1), history.history['loss'], label='Training Loss')
plt.plot(range(1, num_epochs + 1), history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)

# Plot training and validation accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, num_epochs + 1), history.history['accuracy'], label='Training Accuracy')
plt.plot(range(1, num_epochs + 1), history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

The graph above illustrates the performance of the training and validation phases over epochs, depicting the losses associated with the model's predictions on genres.
